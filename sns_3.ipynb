```python
import os
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

print("=====================")
print("pixabay 이미지 크롤러")
print("=====================")
# 크롤링할 키워드와 건수 설정
keyword = input('1. 크롤링할 키워드는?(예:고양이): ')
num = int(input('2. 크롤링할 건수는?: '))
base_dir = input("3.결과 저장할 폴더명(예:c:\\data\\):")

count = 0
pagi = 1

# 결과 파일 저장 폴더 생성
current_time = time.strftime("%Y-%m-%d-%H-%M-%S", time.localtime())
save_dir = os.path.join(base_dir, current_time + "-" + keyword)
os.makedirs(save_dir, exist_ok=True)

# 웹 드라이버 설정
driver_path = "C:\py_temp\chromedriver\chromedriver.exe"
driver = webdriver.Chrome(driver_path)
driver.get("https://pixabay.com/ko/")
time.sleep(2)

# 검색어 입력
search_input = driver.find_element(By.NAME, 'search')
search_input.click()
search_input.send_keys(keyword)
search_input.submit()
time.sleep(2)

# 이미지 크롤링
while count < num:
    time.sleep(2)
    
    body = driver.find_element(By.CSS_SELECTOR, 'body')
    current_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == current_height:
            break
        current_height = new_height
    
    full_html = driver.page_source
    soup = BeautifulSoup(full_html, 'html.parser')
    imgs = soup.find_all('img')
    
    for img in imgs:
        time.sleep(1)
        try:
            img_url = img.get('src')
            f_name = str(count) + ".jpg"
            response = requests.get(img_url)
            with open(os.path.join(save_dir, f_name), 'wb') as f:
                f.write(response.content)
            count += 1
            print(str(count) + "번째 이미지 다운로드.")
            if count == num:
                break
        except:
            print("이미지 다운로드 실패")
    
    if count < num:
        pagi += 1
        next_pagi = "https://pixabay.com/ko/images/search/" + keyword + "/?pagi=" + str(pagi)
        driver.get(next_pagi)

print("추출에 성공했습니다.")
```
