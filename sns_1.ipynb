```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import sys
import numpy
import pandas as pd
import os
import xlwt

keyword  = input('크롤링할 키워드는 무엇입니까?:')
from_d =  input('조회 시작일자를 입력하세요(예:2023/01/01):')
to_d = input('조회 종료일자를 입력하세요(예:2023/12/31):')
f_dir = input("3.결과 파일을 저장할 폴더명만 쓰세요(예:c:\\data\\):")

n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
os.makedirs(f_dir+s+'-'+keyword)
os.chdir(f_dir+s+'-'+keyword)
f_txt=f_dir+s+'-'+keyword+'\\'+s+'-'+keyword+'.txt'
f_csv=f_dir+s+'-'+keyword+'\\'+s+'-'+keyword+'.csv'
f_xls=f_dir+s+'-'+keyword+'\\'+s+'-'+keyword+'.xls'

path = "C:\py_temp\chromedriver\chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get('https://www.g2b.go.kr/index.jsp')
time.sleep(2)

search = driver.find_element(By.ID,"bidNm")
search.send_keys(keyword)

from_date= driver.find_element(By.ID,"fromBidDt")
from_date.clear()
from_date.send_keys(from_d)

to_date = driver.find_element(By.ID,"toBidDt")
to_date.clear()
to_date.send_keys(to_d)

driver.find_element(By.CLASS_NAME,'btn_dark').click()
time.sleep(2)

count = 1
cnt_list = []
c1 = []
c2 = []
c3 = []
c4 = []
c5 = []
c6 = []
c7 = []
c8 = []
c9 = []
c10 = []
c11 = []

driver.switch_to.frame('sub')
driver.switch_to.frame('main')

table = driver.find_element(By.CLASS_NAME, 'table_list_tbidTbl')
tbody = table.find_element(By.TAG_NAME, 'tbody')
rows = tbody.find_elements(By.TAG_NAME,'tr')

for index, row in enumerate(rows):
    content = row.find_elements(By.TAG_NAME, 'td')
    
    cnt_list.append(count)
    print('번호:', count)

    c_1 = content[0].text
    c1.append(c_1)
    print('용역:', c_1)

    c_2 = content[1].text
    c2.append(c_2)
    print('공고번호-차수:', c_2)
   
    c_3 = content[2].text
    c3.append(c_3)
    print('분류:',c_3)
    
    c_4_5 = content[3].find_element(By.TAG_NAME, 'div').find_element(By.TAG_NAME, 'a')
    c4.append(c_4_5.text)
    print('공고명:', c_4_5.text)
    
    c_5 = c_4_5.get_attribute("href")
    c5.append(c_5)
    print('URL 주소:',c_5)
    
    c_6 = content[4].text
    c6.append(c_6)
    print('공고기관:',c_6)
   
    c_7 = content[5].text
    c7.append(c_7)
    print('수요기관:',c_7)
    
    c_8 = content[6].text
    c8.append(c_8)
    print('계약방법:',c_8)
    
    c_9 = content[7].text
    c9.append(c_9)
    print('입력일시:',c_9)
    
    c_10 =content[8].text
    c10.append(c_10)
    print('공동수급:',c_10)
   
    c_11 = content[9].text
    c11.append(c_11)
    print('통찰여부:',c_11)
    
    print('\n\n\n')

    count += 1
        
    time.sleep(2)

resultFrame = pd.DataFrame()
resultFrame['용역']=c1
resultFrame['공고번호-차수']=c2
resultFrame['분류']=c3
resultFrame['공고명']=c4
resultFrame['URL']=c5
resultFrame['공고기관']=c6
resultFrame['수요기관']=c7
resultFrame['계약방법']=c8
resultFrame['입력일시']=c9
resultFrame['공동수급']=c10
resultFrame['통찰여부']=c11

resultFrame.to_csv(f_csv, encoding="utf-8-sig") 
resultFrame.to_excel(f_xls)

f = open(f_txt, 'a',encoding='UTF-8')

for i in range(len(cnt_list)):
    f.write(str(cnt_list[i]) + '번째 공고내용 추출합니다. ~~~~\n')
    f.write('1.업무:' + c1[i] + '\n')
    f.write('2.공고번호-차수:' + c2[i] + '\n')
    f.write('3.분류:' + c3[i] + '\n')
    f.write('4.공고명:' + c4[i] + '\n')
    f.write('5.URL:' + c5[i] + '\n')
    f.write('6.공고기관:' + c6[i] + '\n')
    f.write('7.수요기관:' + c7[i] + '\n')
    f.write('8.계약방법:' + c8[i] + '\n')
    f.write('9.입력일시(입찰마감일시):' + c9[i] + '\n')
    f.write('10.공동수급:' + c10[i] + '\n')
    f.write('11.투찰여부:' + c11[i] + '\n')
        
    f.write('\n**********************************************************\n\n')

f.close()
print("추출 성공했습니다.")
```
