```python
#Step 1. 필요한 모듈과 라이브러리를 로딩합니다.
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import re
import urllib.request  # urllib 모듈을 import합니다.

#Step 2. 사용자에게 주소를 입력 받습니다.
keyword = input('1. 크롤링할 키워드는 무엇입니까?: ')
#keyword = '빅데이터분석'
cnt = int(input('2. 위 주제로 댓글을 크롤링할 유튜브 영상은 몇건입니까?: '))
#cnt = 3
f_dir = input('3. 파일이 저장될 경로만 쓰세요(예: c:\\temp\\ ')
#f_dir = "C:\\py_temp\\test\\"

if  os.path.isdir(f_dir) :
    print('입력하신 경로가 존재하여  {} 폴더에 저장하겠습니다'.format(f_dir))
else :       
    os.makedirs(f_dir)
    print('입력하신 경로가 존재하지 않아 [] 폴더를 생성했습니다'.format(f_dir))

s_time = time.time( )

options = webdriver.ChromeOptions()
    
driver = webdriver.Chrome('C:\py_temp\chromedriver\chromedriver.exe')  # 크롬 드라이버 경로를 지정하세요

page_cnt = 1

print("\n")
print("요청하신 데이터를 수집 중이오니 잠시만 기다려 주세요~~^^")

driver.get('https://www.google.com/search?q={}'.format(keyword))

time.sleep(2)

for page_cnt in range(1, cnt + 1):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 첫 번째 자식 a 요소의 href 속성 값들을 리스트로 담기
    div_elements = soup.find_all('div','yuRUbf')
    
    href_list = []
    for div_element in div_elements:
        first_a_element = div_element.find('a')
        if first_a_element:
            href_list.append(first_a_element['href'])

download_count = 0 
        # 각 링크를 클릭하고 PDF 파일을 저장합니다.
for index, link in enumerate(href_list, start=1):
    if download_count >= cnt:
        break
    time.sleep(2)
    try:
        time.sleep(2)
        # PDF 파일 다운로드 링크를 찾습니다.
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        download_link = soup.find('a', href=re.compile(r'\.pdf$'))
        if download_link:
            file_url = download_link['href']
            print(file_url)
            filename = '{}_{}.pdf'.format(keyword, index)
            urllib.request.urlretrieve(file_url, os.path.join(f_dir, filename))  # urllib.request 모듈을 사용하여 PDF 파일을 다운로드합니다.
            print('{} 파일 저장 완료'.format(filename))
            download_count += 1  # 다운로드한 파일 수 증가
        else:
            print('PDF 다운로드 링크를 찾을 수 없습니다. 페이지가 넘어갑니다.')

    except Exception as e:
        print('오류 발생:', str(e))

    time.sleep(2)  # 파일이 다운로드되기를 기다립니다..
    next_button = driver.find_element(By.XPATH, '//a[@id="pnnext"]')
    next_button.click()
    time.sleep(2)  # 페이지가 로드되기를 기다립니다.
    
    
time.sleep(2)

e_time = time.time( )
t_time = e_time - s_time

print("=" *100)
print("총 소요시간은 %s 초 입니다 " %round(t_time,1))
print("파일이 저장된 폴더명 : %s " %f_dir)
print("=" *100)

driver.close( )
```
